```python
import pandas as pd
import random
from faker import Faker

# Initialize Faker for generating fake data
fake = Faker()

# Create an empty DataFrame
df = pd.DataFrame(columns=["ID", "First_Name", "Last_Name", "DOB", "Address", "Email", "latitude", "longitude", "country"])

# Generate 1000 rows of data
for _ in range(1000):
    row = {
        "ID": fake.unique.random_number(digits=6),
        "First_Name": fake.first_name(),
        "Last_Name": fake.last_name(),
        "DOB": fake.date_of_birth(minimum_age=18, maximum_age=80),
        "Address": fake.address(),
        "Email": fake.email(),
        "latitude": round(random.uniform(-90, 90), 6),
        "longitude": round(random.uniform(-180, 180), 6),
        "country": fake.country()
    }
    df = df.append(row, ignore_index=True)

# Save the dataset to a CSV file
df.to_csv("sample_dataset.csv", index=False)
print("Dataset created and saved as 'sample_dataset.csv'")
```

```python
from mimesis import Generic, Datetime
import pandas as pd

# Initialize Mimesis for German data (you can choose any other locale)
mimesis_data = Generic("de")
datetime = Datetime(locale="de")

# Generate 1000 rows of data
df = pd.DataFrame(columns=["ID", "First_Name", "Last_Name", "DOB", "Address", "Email", "latitude", "longitude", "country"])

for _ in range(4000):  # Extend to 5000 rows
    row = {
        "ID": mimesis_data.random.randint(100000, 999999),
        "First_Name": mimesis_data.person.first_name(),
        "Last_Name": mimesis_data.person.last_name(),
        "DOB": datetime.date(start=1944, end=2006),  # Adjust range for age
        "Address": mimesis_data.address.address(),
        "Email": mimesis_data.person.email(),
        "latitude": mimesis_data.address.latitude(),
        "longitude": mimesis_data.address.longitude(),
        "country": mimesis_data.address.country()
    }
    df = df.append(row, ignore_index=True)

# Save the extended dataset to a CSV file
df.to_csv("extended_sample_dataset.csv", index=False)
print("Extended dataset created and saved as 'extended_sample_dataset.csv'")
```
